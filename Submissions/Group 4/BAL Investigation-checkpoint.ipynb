```python
import numpy as np
import matplotlib.pyplot as plt
```

Load in the data of all the quasars from the file.

```python
quasars=np.loadtxt('sdssagn.txt', delimiter=',')
```

Split the data into separate columns to work with and create an array of numbers 1 to 1000 to plot as the number of the quasar.

```python
redshift=quasars[:,2]
bhmass=quasars[:,3]
luminosity=quasars[:,4]
quasar_number=np.linspace(1,1000,1000)
```

Plot the number of quasars against their redshifts to try and visualise the threshold to classify BALs

```python
plt.scatter(quasar_number,redshift,color='black',marker='o',alpha=0.5)
#plt.axhline(2.1,color='r')
plt.xlabel('Quasar Number')
plt.ylabel('Redshift')
```

Classify the quasars as either BALs or Non-BALs based on their redshifts

```python
redbal=[]
balnumber=[]
rednonbal=[]
nonbalnumber=[]
for i in range(len(redshift)):
    if quasars[:,2][i]>2.1:
        redbal.append(redshift[i])
        balnumber.append(i)
    else:
        rednonbal.append(redshift[i])
        nonbalnumber.append(i)
```

Colour coding the divide between BALs and Non-BALs

```python
plt.scatter(balnumber,redbal,color='red',marker='o',alpha=0.5,label='BAL')
plt.scatter(nonbalnumber,rednonbal,color='b',marker='o',alpha=0.5,label='Non-BAL')
plt.xlabel('Quasar Number')
plt.ylabel('Redshift')
plt.legend()
plt.savefig('BALClassification.jpeg')
```

Getting the masses of the quasars to plot

```python
massbal=[]
balnumber=[]
massnonbal=[]
nonbalnumber=[]
for i in range(len(redshift)):
    if quasars[:,2][i]>2.1:
        massbal.append(bhmass[i])
        balnumber.append(i)
    else:
        massnonbal.append(bhmass[i])
        nonbalnumber.append(i)
```

Plotting the masses of the quasars against their number

```python
plt.scatter(balnumber,massbal,color='red',marker='o',alpha=0.5,label='BAL')
plt.scatter(nonbalnumber,massnonbal,color='b',marker='o',alpha=0.5,label='Non-BAL')
plt.xlabel('Quasar Number')
plt.ylabel('Mass of Black Hole (Log $M_{\odot}$)')
plt.legend(loc='lower right')
plt.savefig('MassComparison.jpeg')
```

Calculating the average mass of BALs and Non-BALs

```python
average_bal_mass=sum(massbal)/len(massbal)
average_nonbal_mass=sum(massnonbal)/len(massnonbal)
print(f'The average mass of a BAL quasar is {average_bal_mass}')
print(f'The average mass of a non BAL quasar is {average_nonbal_mass}')
```

Getting the luminosity of the quasars

```python
lumibal=[]
balnumber=[]
luminonbal=[]
nonbalnumber=[]
for i in range(len(redshift)):
    if quasars[:,2][i]>2.1:
        lumibal.append(luminosity[i])
        balnumber.append(i)
    else:
        luminonbal.append(luminosity[i])
        nonbalnumber.append(i)
```

Plotting the luminosities of the quasars against their number

```python
plt.scatter(balnumber,lumibal,color='red',marker='o',alpha=0.5,label='BAL')
plt.scatter(nonbalnumber,luminonbal,color='b',marker='o',alpha=0.5,label='Non-BAL')
plt.xlabel('Quasar Number')
plt.ylabel('Luminosity (Log W)')
plt.legend()
plt.savefig('LuminosityComparison.jpeg')
```
